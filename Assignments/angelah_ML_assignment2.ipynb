### Question 9

In this exercise, we will predict the number of applications received
using the other variables in the College data set.


```python
college=load_data('College')
```

(a) Split the data set into a training set and a test set.


```python
college_train,college_test = train_test_split(college,test_size=0.6,random_state=0)
college_train['Private01']=np.where(college_train['Private']=='Yes',1,0)
college_test['Private01']=np.where(college_test['Private']=='Yes',1,0)
```

(b) Fit a linear model using least squares on the training set, and report the test error obtained.


```python
design = MS(college.columns.drop(['Apps','Private'])).fit(college)
X_train = design.transform(college_train)
y_train=college_train['Apps']
result=OLS(y_train,X_train).fit()
result.summary()
```

```python
feature_cols = college.columns.drop(['Apps', 'Private']).tolist()
X_train = college_train[feature_cols]
X_train = sm.add_constant(X_train)
y_train = college_train['Apps']
result = sm.OLS(y_train, X_train).fit()

X_test = college_test[feature_cols]
X_test = sm.add_constant(X_test)
y_test = college_test['Apps']


test_predictions = result.predict(X_test)

mse_ols = mean_squared_error(y_test, test_predictions)
rmse_ols = np.sqrt(mse_ols)

print(f"Mean Squared Error (MSE) on Test Set: {mse_ols:,.2f}")
print(f"Root Mean Squared Error (RMSE) on Test Set: {rmse_ols:,.2f}")
```

(c) Fit a ridge regression model on the training set, with λ chosen by cross-validation. Report the test error obtained.


```python
numerical_features = X_train.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_train.select_dtypes(include='object').columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

alphas = np.logspace(-2, 4, 100)

ridge_cv_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RidgeCV(
        alphas=alphas,
        cv=10,
        scoring='neg_mean_squared_error'
    ))
])

ridge_cv_model.fit(X_train, y_train)

best_alpha = ridge_cv_model.named_steps['regressor'].alpha_
test_predictions = ridge_cv_model.predict(X_test)
mse_ridge = mean_squared_error(y_test, test_predictions)
rmse_ridge = np.sqrt(mse_ridge)

print(f"The best alpha: {best_alpha}")
print(f"Mean Squared Error (MSE) on Test Set: {mse_ridge:,.2f}")
print(f"Root Mean Squared Error (RMSE) on Test Set: {rmse_ridge:,.2f}")
```

(d) Fit a lasso model on the training set, with λ chosen by cross-
validation. Report the test error obtained, along with the number of non-zero coefficient estimates.


```python
lasso_cv_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LassoCV(
        alphas=alphas,
        cv=10,
        max_iter=50000,
        random_state=42,
        selection='random'
    ))
])

lasso_cv_model.fit(X_train, y_train)

test_predictions = lasso_cv_model.predict(X_test)
mse_lasso = mean_squared_error(y_test, test_predictions)
rmse_lasso = np.sqrt(mse_lasso)

coefficients = lasso_cv_model.named_steps['regressor'].coef_

tolerance = 1e-4
non_zero_coefficients = np.sum(np.abs(coefficients) > tolerance)

# --- 4. Print Results ---
print(f"Best alpha: {lasso_cv_model.named_steps['regressor'].alpha_:.4f}")
print(f"Mean Squared Error (MSE) on Test Set: {mse_lasso:,.2f}")
print(f"Root Mean Squared Error (RMSE) on Test Set: {rmse_lasso:,.2f}")
print(f"Number of Non-Zero Coefficient Estimates: {non_zero_coefficients}")
```

(e) Fit a PCR model on the training set, with M chosen by cross-
validation. Report the test error obtained, along with the value
of M selected by cross-validation.


```python
X_train_processed = preprocessor.fit_transform(X_train)
n_features = X_train_processed.shape[1]

M_range = np.arange(1, n_features + 1)

pcr_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('regressor', LinearRegression())
])

param_grid = {
    'pca__n_components': M_range
}

grid_search = GridSearchCV(
    pcr_pipe,
    param_grid,
    cv=10,
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

best_pcr_model = grid_search.best_estimator_

test_predictions = best_pcr_model.predict(X_test)
mse_pcr = mean_squared_error(y_test, test_predictions)
rmse_pcr = np.sqrt(mse_pcr)


best_M = grid_search.best_params_['pca__n_components']

print(f"Best M: {best_M}")
print(f"Mean Squared Error (MSE) on Test Set: {mse_pcr:,.2f}")
print(f"Root Mean Squared Error (RMSE) on Test Set: {rmse_pcr:,.2f}")
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
n_features = X_train_processed.shape[1]

M_range = np.arange(1, n_features + 1)

pls_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', PLSRegression())
])

param_grid = {
    'regressor__n_components': M_range
}

grid_search = GridSearchCV(
    pls_pipe,
    param_grid,
    cv=10,
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

best_pls_model = grid_search.best_estimator_

test_predictions = best_pls_model.predict(X_test)
mse_pls = mean_squared_error(y_test, test_predictions)
rmse_pls = np.sqrt(mse_pls)

best_M = grid_search.best_params_['regressor__n_components']

print(f"Best M: {best_M}")
print(f"Mean Squared Error (MSE) on Test Set: {mse_pls:,.2f}")
print(f"Root Mean Squared Error (RMSE) on Test Set: {rmse_pls:,.2f}")
```

```python
College_train, College_valid = train_test_split(College, test_size=0.7, random_state=0)
```

```python
College['Private']=College['Private'].astype('category')
College_train['Private']=College_train['Private'].astype('category')
College_valid['Private']=College_valid['Private'].astype('category')
y=College_train['Outstate']
Vars_train = College_train.columns.drop(['Outstate'])
design=MS(Vars_train).fit(College_train)
X_train=design.transform(College_train)
College_train.info()
```

```python
def nCp(sigma2, estimator, X, Y):
    "Negative Cp statistic"
    n, p = X.shape
    Yhat = estimator.predict(X)
    RSS = np.sum((Y - Yhat)**2)
    return -(RSS + 2 * p * sigma2) / n
design_sigma = MS(College.columns.drop('Outstate')).fit(College)
Y_sigma = np.array(College['Outstate'])
X_sigma = design_sigma.transform(College)
sigma2 = OLS(Y_sigma,X_sigma).fit().scale
neg_Cp = partial(nCp, sigma2)
```

```python
strategy=Stepwise.first_peak(design,direction='forward',max_terms=len(design.terms))
Outstate_MSE = sklearn_selected(OLS,strategy,scoring=neg_Cp)
Outstate_MSE.fit(College_train, y)
Outstate_MSE.selected_state_
```

Using forward stepwise selection and optimizing for the Cp coefficient, a model with 7 predictors was identified as the best fit.

```python
np.random.seed(1)
X = np.random.rand(60,50)
X[:20,:20] += 1
X[20:40,20:40] -= 1

y=np.concatenate((np.repeat(0,20),np.repeat(1,20),np.repeat(2,20)))
```

```python
plt.scatter(pca_X[:,0], pca_X[:,1], c=y, cmap=mpl.cm.Paired)
plt.xlabel('Z1')
plt.ylabel('Z2');
```

(c) Perform K-means clustering of the observations with K = 3. How well do the clusters that you obtained in K-means clustering compare to the true class labels?

```python
kmeans_3C = KMeans(n_clusters=3,random_state=2,n_init=20).fit(X)
pd.crosstab(index=kmeans_3C.labels_,columns=y,rownames=['kmeans_3C'],colnames=['y'])
```
